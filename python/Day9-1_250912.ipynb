```python
# PyQT5로 그리드 레이아웃 만들기

from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import (QWidget, QApplication, QLabel, QLineEdit, QTextEdit, QGridLayout)
import sys

class MyApp(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()
        
    def initUI(self):
        # 초기화면 구성
        grid = QGridLayout() # 내가 사용할 grid layout 생성
        self.setLayout(grid) # 내가 만든 grid Layout을 제일 바깥 Layout으로 설정

        grid.addWidget(QLabel("Title : "), 0, 0)
        grid.addWidget(QLabel("Author : "), 1, 0)
        grid.addWidget(QLabel("Review : "), 2, 0)

        grid.addWidget(QLineEdit(), 0, 1)
        grid.addWidget(QLineEdit(), 1, 1)
        grid.addWidget(QTextEdit(), 2, 1)
           
        self.setWindowTitle("grid layout")
        self.setGeometry(300, 300, 300, 200)
        
        
        # setStyleSheet에서 자주쓰는 명령어
        # 배경색(background-color:), 글자색(color:)
        # 폰트 설정
        # 글꼴(font-family: Arial), 글자크기(font-size:16px;), 굵기(font-weight: nomal, bold)
        # 테두리(border: 2px solid black), 둥근모서리(border-radius: px)
        # 패딩(padding)_안쪽 여백_ 과 마진(margin)_바깥 여백_
        # 호버(hover) : 가상 클래스 사용 가능 _마우스 올렸을 때 효과_
        # 위젯별 선택자 (QpushButton,Qlabel,QlineEdit 등..) :focus는 포커스 있을 때
        # 글자 정렬 (Qt속성 스타일로 텍스트 정렬가능 : AlignLeft, AlignCenter, AlignRight)
        # 투명도 설정 (rgba 0~255, 255, 0, 0, x 앞에 3곳 색상 넣고 마지막 X값이 투명도)
        
        self.setStyleSheet("""  
        
        QLabel {
            color:#0000FF;
        }        
        QLineEdit {
            font-size: 14px;
            border: 1px solid #FF0000;
        }
        """)
                
        self.show()


if __name__ == "__main__": # 내가 직접 실행한다면
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    exam = MyApp()
    sys.exit(app.exec_())

```

```python
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import (QWidget, QApplication, QLabel, QLineEdit, QTextEdit,
                            QGridLayout, QPushButton)
import sys

class MyApp(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()
        
    def initUI(self):
        # 초기화면 구성
        grid = QGridLayout() # 내가 사용할 grid layout 생성
        self.setLayout(grid) # 내가 만든 grid Layout을 제일 바깥 Layout으로 설정
        self.result = ""
        
        

        value_list = ["7", "8", "9", "*",
                     "4", "5", "6", "-",
                     "1", "2", "3", "+",
                     "C", "0", "=", "/"]
      
        button_list = []
        for i in range(16):
            button = QPushButton()
            button.clicked.connect(self.button_cliked)
            button_list.append(button)

        for i, button in enumerate(button_list):
            button.setText(value_list[i])
            

        self.ql_result = QLabel("0", alignment=Qt.AlignRight)
        grid.addWidget(self.ql_result, 0, 0, 1, 4)
        # addWidget (행, 열, 차지할 행, 차지할 열)

        
        count = 0
        for i in range(1, 5):
            for j in range(4):
                grid.addWidget(button_list[count], i, j)
                count += 1
        
            
        
        

        
        
        self.setLayout(grid)
        
        self.setWindowTitle("계산기")
        self.setGeometry(300, 300, 300, 200)
        

        self.setStyleSheet("""
        
       QWidget {
                background-color: #282c34;
            }
            QLabel {
                font-size: 40px;
                color: #ffffff;
                padding: 15px;
                border: 2px solid #61afef;
                border-radius: 10px;
                background-color: #21252b;
            }
            QPushButton {
                font-size: 24px;
                color: #ffffff;
                background-color: #61afef;
                border: none;
                border-radius: 10px;
                padding: 15px;
                margin: 5px;
                font-weight: bold;
            }
            QPushButton:hover {
                background-color: #528ecc;
            }
            QPushButton:pressed {
                background-color: #3b6ea5;
            }
        """)
        
      
        self.show()

        
    def button_cliked(self):
        # __init__, initUI()와 같은 level
        # blah blah blah~~~~~
        # 버튼 3을 눌렀다 -> QLabel의 text가 3이 되게
        # button clicked 나를 부른 사람이 누구야
        button = self.sender() # 버튼 3 누르면 3버튼
        key = button.text()
        
        if key == "=":
            try:
                cal_result = eval(self.result)
                self.ql_result.setText(str(cal_result))
                self.result = ""
            except Exception:
                self.ql_result.setText("Error")
                self.result = ""
        elif key == "C":
            self.result = ""
            self.ql_result.setText("0")
        else:
            self.result += key
            self.ql_result.setText(self.result)
            

if __name__ == "__main__": # 내가 직접 실행한다면
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    exam = MyApp()
    sys.exit(app.exec_())
```

```python
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import (
    QWidget, QApplication, QLabel, QGridLayout,
    QPushButton, QVBoxLayout
)
from PyQt5.QtGui import QFont
import sys
import random
import time

class MyApp(QWidget):
    def __init__(self):
        super().__init__()
        self.n = 5           # 5x5 게임 (필요시 조절)
        self.n2 = self.n ** 2
        self.initUI()

    def initUI(self):
        self.answer = 1
        self.setWindowTitle(f"1 to {self.n2} - POKEMON STYLE")
        self.setGeometry(300, 300, 600, 700)

        vbox = QVBoxLayout()

        # 결과 표시 라벨
        self.ql_result = QLabel(f"1 to {self.n2}", alignment=Qt.AlignCenter)
        self.ql_result.setFont(QFont("Arial", 24))
        vbox.addWidget(self.ql_result)

        # 스타트 버튼
        self.button_start = QPushButton("START BATTLE!")
        self.button_start.setObjectName("StartButton")
        self.button_start.setFixedHeight(50)
        self.button_start.clicked.connect(self.start_game)
        vbox.addWidget(self.button_start)

        # 버튼 그리드
        self.grid = QGridLayout()
        self.button_list = []

        for i in range(self.n2):
            button = QPushButton()
            button.setMinimumSize(80, 60)
            button.setStyleSheet("font-size:20px; font-weight:bold;")
            button.setEnabled(False)
            self.button_list.append(button)

        count = 0
        for i in range(self.n):
            for j in range(self.n):
                self.grid.addWidget(self.button_list[count], i, j)
                count += 1

        vbox.addLayout(self.grid)
        self.setLayout(vbox)

        # 스타일 시트 (포켓몬 느낌!)
        self.setStyleSheet("""
            QWidget {
                background-color: #A7DB8D; /* 풀밭 느낌 */
            }

            QLabel {
                color: #3B3B3B;
                font-size: 32px;
                padding: 10px;
                border: 3px solid #7AC74C;
                border-radius: 10px;
                background-color: #E6F2D8;
            }

            QPushButton {
                font-size: 18px;
                border-radius: 10px;
                padding: 10px;
                color: white;
            }

            QPushButton:hover {
                background-color: #FFD700;
                color: black;
            }

            QPushButton:pressed {
                background-color: #FF6F61;
            }

            QPushButton#StartButton {
                background-color: #EF5350; /* 몬스터볼 빨강 */
                color: white;
                font-size: 20px;
                font-weight: bold;
                border: 3px solid #3B3B3B;
            }

            QPushButton#StartButton:hover {
                background-color: #D32F2F;
            }
        """)

        self.show()

    def start_game(self):
        self.button_start.setEnabled(False)
        number_list = list(range(1, self.n2 + 1))
        random.shuffle(number_list)

        poke_colors = ['#F7D02C', '#6890F0', '#EE8130', '#A33EA1', '#7AC74C', '#F95587']
        for i in range(self.n2):
            self.button_list[i].setEnabled(True)
            self.button_list[i].setText(str(number_list[i]))

            # 포켓몬 타입 컬러 랜덤 적용
            color = random.choice(poke_colors)
            self.button_list[i].setStyleSheet(f"""
                QPushButton {{
                    background-color: {color};
                    font-size: 20px;
                    font-weight: bold;
                    color: white;
                    border: 2px solid #444;
                    border-radius: 10px;
                }}
            """)

            self.button_list[i].clicked.connect(self.button_clicked)

        self.start_time = time.perf_counter()

    def button_clicked(self):
        button = self.sender()
        if int(button.text()) == self.answer:
            button.setEnabled(False)
            button.setText("")
            self.answer += 1

            if self.answer > self.n2:
                self.end_time = time.perf_counter()
                record_time = self.end_time - self.start_time
                self.ql_result.setText(f"🎉 {record_time:.2f} sec")
                self.button_start.setEnabled(True)
                self.answer = 1


if __name__ == "__main__":
    app = QApplication(sys.argv)
    ex = MyApp()
    sys.exit(app.exec_())
```
